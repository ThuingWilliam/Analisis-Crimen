libreria que utilizaremos en analisis pandas

```python
import pandas as pd
from rich import print
from rich.table import Table
import jinja2
import folium
```

```python
data = pd.read_csv(r'Crimen.csv')
data.head(10)
```

Aqui Mostramos la Cantidad de Fila de Nuestro DataFrame

```python
cantidad=len(data)

cantidaddata=Table(title="Cantidad de Datos de nuestra Muestra".upper())
cantidaddata.add_column("Cantidad",justify='center',style="")
cantidaddata.add_row(str(cantidad))

cantidaddata
```

Como Podemos ver del data Frame Principal sacamos ciertas columna y la renombramos para poder hacer un analisis mas estructurado


```python
data.rename(columns={'Date Rptd':'Fecha'},inplace=True)
data.rename(columns={'Rpt Dist No':'Distrito'},inplace=True)
data.rename(columns={'Vict Age':'Victima edad'},inplace=True)
data.rename(columns={'Vict Sex':'Victima Sexo'},inplace=True)
data.rename(columns={'Crm Cd Desc':'Tipo de Crimen'},inplace=True)
data.rename(columns={'AREA':'Cod Area'},inplace=True)
data.rename(columns={'AREA NAME':'Nombre del Area'},inplace=True)
data.rename(columns={'LOCATION':'Ubicacion'},inplace=True)
table=data[['Fecha','Cod Area','Nombre del Area','Distrito','Ubicacion','Victima edad','Victima Sexo','Tipo de Crimen']]
table.head(10)
```

```python
media_edad = table['Victima edad'].mean()
print("la media de las victima fue {:.2f}".format(media_edad,2))
```

```python
maximaedad = table['Victima edad'].max()
print("la edad maxima de una victima {:.0f}".format(maximaedad,2))
```

```python
minima_edad = table['Victima edad'][table['Victima edad'] > 0].min()
print("La edad mínima de una víctima {:.2f} años.".format(minima_edad))
```

Aqui mostramos una Agrupacion por Sexo la Cantidad de Victima


```python
Vsexo=table.groupby('Victima Sexo').size().sort_values(ascending=False)
sexotable=Table(title="Cantidad de Victima por Sexo",style="#97efe9",title_justify="center",title_style="#97efe9")
sexotable.add_column("Sexo")
sexotable.add_column("Cantidad de Victima",justify='center')
for sexo,conteo in Vsexo.items():
    sexotable.add_row(str(sexo),str(conteo))
sexotable.add_row("Total",str(Vsexo.sum()))
sexotable
```

Aqui estamos Mostrando un breve Analisis de las Victimas por Años


```python
data.rename(columns={'Date Rptd':'Fecha'}, inplace=True)
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%d/%m/%Y')
data['year'] = data['Fecha'].dt.year
# Agrupar por años y contar el número de ocurrencias (víctimas)
cyear = data.groupby('year').size().sort_values(ascending=False)
table_richano = Table(title="Cantidad de Víctimas por Años Desde 2020-2024", style="#e83151",title_style="#d2cca1")
table_richano.add_column("Años", style="",justify="center")
table_richano.add_column("Víctimas", justify="center")
for a, conteo in cyear.items():
    table_richano.add_row(str(a), str(conteo))
table_richano.add_row("Total",str(cyear.sum()))
table_richano
```

Ahora veremos el top 10 de Crimen en EEUU




```python
top10crime = table['Tipo de Crimen'].value_counts().head(10)
valor = 0
table_rich = Table(title="Los 10 Crímenes Más Comunes EEUU",style="#97efe9",title_style="#97efe9")
# Agregar columnas
table_rich.add_column("No.", style="bold White")
table_rich.add_column("Crimen", style="White ",justify='center')
table_rich.add_column("Cantidad", style="bold yellow",justify='center')
# Llenar la tabla con los datos
for nombre, cantidad in top10crime.items():
    valor += 1
    table_rich.add_row(str(valor), nombre, str(cantidad))
table_rich
```

Veremos la cantidad de Victima por Años y su Sexo


```python
anospivot = table.pivot_table(
    index=data['year'],
    columns=data['Victima Sexo'],
    aggfunc='size'
)
anospivot['Total'] = anospivot.sum(axis=1)
anospivot_sorted = anospivot.sort_index(ascending=False)
fpivotano = anospivot_sorted.style.format({
    'Total': '{:,.0f}',
    'F': '{:,.0f}',
    'M': '{:,.0f}',

})
fpivotano
```

Aqui Mostraremos top 100 de los lugares donde hubo Victimas

```python

data.rename(columns={'LAT':'latitud'},inplace=True)
data.rename(columns={'LON':'longitud'},inplace=True)

data.head(1)

mapa=folium.Map(location=[data['latitud'].mean(),data['longitud'].mean()],zoom_start=10)


for index,row in data.head(100).iterrows():
    folium.Marker(
        location=[row['latitud'],row['longitud']],
    ).add_to(mapa)
mapa
```
